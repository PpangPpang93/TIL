#### 데이터의 구성
    total_bill : 총 결제금액
    tip : 팁
    sex : 성별
    smoker : 흡연여부
    day : 요일
    time : 점심, 저녁 구분
    size : 함께한 인원

**1. 총 결제금액과 팁의 분포를 살펴보기위해 plt의 2 X 1 subplot을 사용하여 첫번째 subplot에는 total_bill 두번째 subplot에는 tip의 히스토그램을 그려봅시다. figsize는 자유롭게 범례(legend)는 데이터의 컬럼명으로 생성합니다**

```python
plt.figure(figsize=(12,8))
plt.subplot(211)
plt.hist(df['total_bill'],label='total_bill')
plt.legend()

plt.subplot(212)
plt.hist(df['tip'],label='tip')
plt.legend()

plt.show()
```

**2. 2 X 2 plt subplot을 생성하고 데이터셋의 성별, 흡연여부, 요일, 점심/저녁구분의 sns의 countplot을 그려봅시다**

```python
plt.figure(figsize=(20,12))

plt.subplot(221)
sns.countplot(df['sex'])

plt.subplot(222)
sns.countplot(df['smoker'])

plt.subplot(223)
sns.countplot(df['day'])

plt.subplot(224)
sns.countplot(df['time'])

plt.show()
```

**3. 어떤 사람들이 팁을 많이 주었을까요? 팁을 8달러 이상 준 사람들의 샘플을 df1으로 저장해봅시다**

```python
df1=df[df['tip'] >= 8]
df1
```

**4. 모든 샘플의 평균과 팁을 많이 준 사람 샘플 평균간의 차이를 한눈에 확인해보고자 합니다. 4 X 2 subplot을 생성하고 1열에는 모든 샘플의 성별, 흡연여부, 요일, 점심/저녁구분 counplot을 그리고 2열에는 팁을 많이 준 사람들의 countplot을 그려봅시다.**

```python
plt.figure(figsize=(20,20))
plt.subplot(421)
sns.countplot(df['sex'])

plt.subplot(423)
sns.countplot(df['smoker'])

plt.subplot(425)
sns.countplot(df['day'])

plt.subplot(427)
sns.countplot(df['time'])


plt.subplot(422)
sns.countplot(df1['sex'])

plt.subplot(424)
sns.countplot(df1['smoker'])

plt.subplot(426)
sns.countplot(df1['day'])

plt.subplot(428)
sns.countplot(df1['time'])


plt.show()
```
