```
import numpy as np
from sklearn.linear_model import LinearRegression
```

```
# reshape 코드입력
```

```
x = np.arange(10).reshape(-1,1)
y = (2*x + 1).reshape(-1,1)
```

```
x
```

```
y
```

## 모델의 선언, fit(학습), predict(예측)

```
# 모델 선언
model = LinearRegression()
```

```
model
```

```
# 학습 mode.fit(데이터, 예측값)
model.fit(x,y) # 위의 결과와 같지만 학습이 완료됨
```

```
# 예측 x에 없는 10을 넣어보고 값을 예측해보기
prediction = model.predict([[10.0]])
```

```
prediction
```

## 용어정리
- feature : x, 학습을 위한 데이서텟, 예측할 값은 빠져 있다
- label : y, 예측해야할 값, 예측값만 있어야한다

- 학습을 위한 데이터 : Training Set, feature, 모델이 학습하기 위해 필요한 데이터. label이 모두 존재 / x_train, y_train
- 예측을 위한 데이터 : Test Set, feature만 존재 / x_test, y_test
